```python
import pandas as pd
import networkx as nx
```

```python
data = pd.read_csv("Pharma.csv")
```

```python
data.isnull().sum()
```

```python
data = data.dropna(subset=["assignee"]).reset_index()
```

```python
a = "inventor_id"
data["Inventor_List"] = None
for index,rows in data.iterrows():
    k = []
    for i in range(1,10):
        if pd.isna(data.at[index,a+str(i)]):
            break
        else:
            k.append(data.at[index,a+str(i)])
    data.at[index,"Inventor_List"] = k
```

```python
data = data[['patent_number','assignee','grant_year','Inventor_List']]
```

```python
data.isnull().sum()
```

```python
data
```

```python
data = data.groupby(['assignee',"grant_year"]).agg({'patent_number':list,"Inventor_List":list}).reset_index()
```

```python
data
```

```python
data["# of Patent"] = data["patent_number"].apply(lambda x : len(x))
data["# of Inventor"] = data["Inventor_List"].apply(lambda x : len(set([item for sublist in x for item in sublist])))
```

```python
data
```

```python
for index,rows in data.iterrows():
    G = nx.Graph()
    edges_lists = data.at[index,"Inventor_List"]
    for sublist in edges_lists:
        for node in sublist:
            G.add_node(node) 
            for other_node in sublist:
                if node != other_node:
                    G.add_edge(node, other_node)  
    data.at[index,"Density"] = nx.density(G)
    data.at[index,"Avg Clustering"] = nx.average_clustering(G)
    data.at[index,"# of Edges"] = G.number_of_edges()
    data.at[index,"# of Nodes"] = G.number_of_nodes()
```

```python
data
```

```python
pharma = data[["assignee","grant_year","# of Patent","# of Inventor","# of Nodes","# of Edges","Density","Avg Clustering"]]
```

```python
pharma.to_csv("Pharma_Patent.csv",index=False)
```
